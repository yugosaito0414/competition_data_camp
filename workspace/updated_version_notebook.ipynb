## üí™ Challenge II
Help your team leader understand your company's products. Include:

1. What is the most expensive item your company sells? The least expensive?
2. How many different products of each category does your company sell?
3. What are the top three brands with the highest average list price? The top three categories?
4. Any other insights you found during your analysis?

1. What is the most expensive item your company sells? The least expensive?

```python
SELECT 
	product_name, list_price
FROM products
where list_price = (select max(list_price) from products)
union all 
SELECT 
	product_name, list_price
FROM products
where list_price = (select min(list_price) from products)
```

2.How many different products of each category does your company sell?

```python
select c.category_id, category_name, sum(product_id) as sold_per_category
from products as p
left join categories as c
on p.category_id = c.category_id
group by c.category_id, category_name
order by sold_per_category desc
```

3.What are the top three brands with the highest average list price? The top three categories?

```python
select b.brand_id, b.brand_name, round(avg(list_price),2) as average_list_price
from products as p
left join brands as b
on p.brand_id = b.brand_id
group by b.brand_id, b.brand_name
order by average_list_price desc
```

4.Any other insights you found during your analysis? I would love to find out the most expensive bikes per year and by brand_name. 

```python
select brand_name, model_year, round(max(list_price),2) as most_expensive_by_brand_and_year
from products as p
left join brands as b
on p.brand_id = b.brand_id
group by brand_name, model_year
order by most_expensive_by_brand_and_year desc
```
