## ğŸ’ª ãƒãƒ£ãƒ¬ãƒ³ã‚¸å•é¡Œ
åŒåƒšãŒCO2æ’å‡ºé‡ã®å°‘ãªã„è»Šä¸¡ã®ç¨®é¡ã«ã¤ã„ã¦åˆ†ã‹ã‚‰ãšå›°ã£ã¦ã„ã‚‹ã‚ˆã†ã§ã™ã€‚ä»¥ä¸‹ã®äº‹ã‚’çŸ¥ã‚ŠãŸã„ã§ã™ï¼š

1. å…¨ã¦ã®è»Šã®ã‚¨ãƒ³ã‚¸ãƒ³ã‚µã‚¤ã‚ºã®ä¸­å¤®å€¤ã¯ä½•ãƒªãƒƒãƒˆãƒ«ã§ã™ã‹?
2. ãƒ¬ã‚®ãƒ¥ãƒ©ãƒ¼ã‚¬ã‚½ãƒªãƒ³ï¼ˆç‡ƒæ–™ç¨®é¡ï¼Xï¼‰ã€ãƒ—ãƒ¬ãƒŸã‚¢ãƒ ã‚¬ã‚½ãƒªãƒ³ï¼ˆç‡ƒæ–™ç¨®é¡ï¼Zï¼‰ã€ã‚¨ã‚¿ãƒãƒ¼ãƒ«ï¼ˆç‡ƒæ–™ç¨®é¡ï¼Eï¼‰ã€è»½æ²¹ï¼ˆç‡ƒæ–™ç¨®é¡ï¼Dï¼‰ã®å¹³å‡ç‡ƒè²»ã¯ã©ã‚Œãã‚‰ã„ã§ã™ã‹ï¼Ÿ 
3. ç‡ƒæ–™æ¶ˆè²»é‡ã¨CO2æ’å‡ºé‡ã®ç›¸é–¢é–¢ä¿‚ã¯ä½•ã§ã™ã‹?
4. SUV-SMALLã¨MID-SIZEã®ã©ã¡ã‚‰ã®è»Šä¸¡ã‚¯ãƒ©ã‚¹ãŒå¹³å‡CO2æ’å‡ºé‡ãŒå°‘ãªã„ã§ã™ã‹?
5. ã™ã¹ã¦ã®è»Šä¸¡ã®å¹³å‡CO2æ’å‡ºé‡ã¯ã©ã‚Œãã‚‰ã„ã§ã™ã‹? ã‚¨ãƒ³ã‚¸ãƒ³ã‚µã‚¤ã‚ºãŒ2.0ãƒªãƒƒã‚¿ãƒ¼ä»¥ä¸‹ã®è»Šä¸¡ã®æ’å‡ºé‡ã«é•ã„ã¯ã‚ã‚Šã¾ã™ã‹ã‹ï¼Ÿ
6. åˆ†æä¸­ã«ä»–ã«ç™ºè¦‹ã—ãŸäº‹ã¯ã‚ã‚Šã¾ã™ã‹?

1. å…¨ã¦ã®è»Šã®ã‚¨ãƒ³ã‚¸ãƒ³ã‚µã‚¤ã‚ºã®ä¸­å¤®å€¤ã¯ä½•ãƒªãƒƒãƒˆãƒ«ã§ã™ã‹?

```python
# å¿…è¦ãŒã‚ã‚‹ã™ã¹ã¦ã®ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import statistics
import math

# å…¨è»Šã®å¹³å‡ã‚¨ãƒ³ã‚¸ãƒ³ã‚µã‚¤ã‚º(L)ã‚’æŠ½å‡º
median_engine_size = statistics.mean(cars_engine_sizes)
print(median_engine_size)
```

2. ãƒ¬ã‚®ãƒ¥ãƒ©ãƒ¼ã‚¬ã‚½ãƒªãƒ³ï¼ˆç‡ƒæ–™ç¨®é¡ï¼Xï¼‰ã€ãƒ—ãƒ¬ãƒŸã‚¢ãƒ ã‚¬ã‚½ãƒªãƒ³ï¼ˆZï¼‰ã€ã‚¨ã‚¿ãƒãƒ¼ãƒ«ï¼ˆEï¼‰ã€è»½æ²¹ï¼ˆDï¼‰ã®å¹³å‡ç‡ƒè²»ã¯ã©ã‚Œãã‚‰ã„ã§ã™ã‹ï¼Ÿ è³ªå•ã«ç­”ãˆã‚‹ã«ã¯ã€cars_fuel_typesã‹ã‚‰Xè¦ç´ ã‚’å–å¾—ã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚

```python
#å…¨è»Šã®å¹³å‡ç‡ƒè²»
all_cars_fuel_average = statistics.mean(cars_fuel_consumption)
print (f"å…¨è»Šã®å¹³å‡ç‡ƒè²»: {round(all_cars_fuel_average, 2)} (L/100 km) ")

#ç‡ƒæ–™ã®ç¨®é¡: X = ãƒ¬ã‚®ãƒ¥ãƒ©ãƒ¼ ã‚¬ã‚½ãƒªãƒ³ã€Z = ãƒ—ãƒ¬ãƒŸã‚¢ãƒ  ã‚¬ã‚½ãƒªãƒ³ã€D = ãƒ‡ã‚£ãƒ¼ã‚¼ãƒ«ã€E = ã‚¨ã‚¿ãƒãƒ¼ãƒ« (E85)ã€N = å¤©ç„¶ã‚¬ã‚¹ã€‚

#ãƒ¬ã‚®ãƒ¥ãƒ©ãƒ¼ã‚¬ã‚½ãƒªãƒ³ã®å¹³å‡ç‡ƒæ–™æ¶ˆè²»é‡
X_cars_fuel_average = df_cars[cars_fuel_types == 'X']['Fuel Consumption Comb (L/100 km)'].mean()
print (f"ãƒ¬ã‚®ãƒ¥ãƒ©ãƒ¼ã‚¬ã‚½ãƒªãƒ³ã®å¹³å‡ç‡ƒè²»: {round(X_cars_fuel_average, 2)} (L/100 km) ")

#ãƒ—ãƒ¬ãƒŸã‚¢ãƒ ã‚¬ã‚½ãƒªãƒ³ã®å¹³å‡ç‡ƒæ–™æ¶ˆè²»é‡
Z_cars_fuel_average = df_cars[cars_fuel_types == 'Z']['Fuel Consumption Comb (L/100 km)'].mean()
print (f"ãƒ—ãƒ¬ãƒŸã‚¢ãƒ ã‚¬ã‚½ãƒªãƒ³ã®å¹³å‡ç‡ƒæ–™æ¶ˆè²»é‡: {round(Z_cars_fuel_average, 2)} (L/100 km) ")

#ãƒ‡ã‚£ãƒ¼ã‚¼ãƒ«ã‚¬ã‚½ãƒªãƒ³ã®å¹³å‡ç‡ƒæ–™æ¶ˆè²»é‡
D_cars_fuel_average = df_cars[cars_fuel_types == 'D']['Fuel Consumption Comb (L/100 km)'].mean()
print (f"ãƒ‡ã‚£ãƒ¼ã‚¼ãƒ«ã‚¬ã‚½ãƒªãƒ³ã®å¹³å‡ç‡ƒæ–™æ¶ˆè²»é‡: {round(D_cars_fuel_average, 2)} (L/100 km) ")

#ã‚¨ã‚¿ãƒãƒ¼ãƒ«ã‚¬ã‚½ãƒªãƒ³ã®å¹³å‡ç‡ƒæ–™æ¶ˆè²»é‡
E_cars_fuel_average = df_cars[cars_fuel_types == 'E']['Fuel Consumption Comb (L/100 km)'].mean()
print (f"ã‚¨ã‚¿ãƒãƒ¼ãƒ«ã‚¬ã‚½ãƒªãƒ³ã®å¹³å‡ç‡ƒæ–™æ¶ˆè²»é‡: {round(E_cars_fuel_average, 2)} (L/100 km) ")

#å¤©ç„¶ã‚¬ã‚¹ç‡ƒæ–™æ¶ˆè²»é‡
N_cars_fuel_average = df_cars[cars_fuel_types == 'N']['Fuel Consumption Comb (L/100 km)'].mean()
print (f"å¤©ç„¶ã‚¬ã‚¹ç‡ƒæ–™æ¶ˆè²»é‡: {round(N_cars_fuel_average, 2)} (L/100 km) ")
```

è³ªå•ã®è§£ç­”ã‹ã‚‰ã€ã‚¨ã‚¿ãƒãƒ¼ãƒ«è»ŠãŒæœ€ã‚‚100kmã‚ãŸã‚Šã§ç‡ƒæ–™ã‚’æ¶ˆè²»ã™ã‚‹ã“ã¨ãŒã‚ã‹ã‚Šã¾ã™ã€‚ æ¬¡ã«å¤©ç„¶ã‚¬ã‚¹ã€3ç•ªç›®ã«ç‡ƒè²»ãŒæ‚ªã„ã®ãŒãƒ—ãƒ¬ãƒŸã‚¢ãƒ ã‚¬ã‚½ãƒªãƒ³ã€‚é€†ã«ç‡ƒè²»ã®æ¶ˆè²»é‡ãŒå°‘ãªã„ã®ã¯ã€ãƒ‡ã‚£ãƒ¼ã‚¼ãƒ«è»Šã§ã™ã€‚ ãƒ‡ã‚£ãƒ¼ã‚¼ãƒ«è»Šã¯ã©ã®è»Šã‚ˆã‚Šã‚‚äºŒé…¸åŒ–ç‚­ç´ ï¼ˆCO2ï¼‰ã®æ¶ˆè²»é‡ãŒå°‘ãªã„ã“ã¨ãŒã‚ã‹ã‚Šã¾ã™ã€‚

3. ç‡ƒæ–™æ¶ˆè²»é‡ã¨CO2æ’å‡ºé‡ã®ç›¸é–¢é–¢ä¿‚ãŒã‚ã‚‹ã‹èª¿ã¹ã¦ä¸‹ã•ã„ã€‚ã‚‚ã—é–¢ä¿‚æ€§ãŒã‚ã‚‹å ´åˆã€æ•£å¸ƒå›³ã‚’ä½¿ã„èª¿ã¹ã¾ã™ã€‚è‡ªå‹•è»ŠãŒç‡ƒæ–™ãŒä½¿ãˆã°ä½¿ã†ã»ã©ã€ã‚ˆã‚Šå¤šãã®CO2ãŒæ’å‡ºã•ã‚Œã‚‹ã¨ä»®å®šã—ã¾ã™ã€‚

```python
import matplotlib.pyplot as plt

#ã‚°ãƒ©ãƒ•ã‚’è¦–è¦šåŒ–ã™ã‚‹ãŸã‚ã«ã€2 ã¤ã®å¤‰æ•°ï¼ˆcars_fuel_consumptionã€cars_co2_emissionsï¼‰ã‚’ä»£å…¥
plt.scatter(cars_fuel_consumption, cars_co2_emissions)
plt.xlabel("cars_fuel_consumption")
plt.ylabel("cars_co2_emissions")
plt.show()

#corrcoefé–¢æ•°ã‚’ä½¿ç”¨ã—ã€2ã¤ã®å¤‰æ•°é–“ãŒæ­£ã®ç›¸é–¢ã‹èª¿ã¹ã¾ã™
np.corrcoef(cars_fuel_consumption, cars_co2_emissions)
```

çµæœã¯äºˆæ¸¬é€šã‚Šã€ç‡ƒè²»ã¨co2æ’å‡ºé‡ã¯æ­£ã®ç›¸é–¢ãŒã‚ã‚Šã¾ã™ã€‚ 

4. ã€ŒSUV-SMALLã€ã¨ã€ŒMID-SIZEã€ã®ã©ã¡ã‚‰ã®è»Šä¸¡ã‚¯ãƒ©ã‚¹ãŒå¹³å‡CO2æ’å‡ºé‡ãŒä½ã„ã§ã™ã‹?

```python
# å¿…è¦ãªåˆ—ã‚’Pandas Seriesã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆã¨ã—ã¦æŠ½å‡º
cars_classes = df_cars['Vehicle Class']
cars_co2_emissions = df_cars['CO2 Emissions(g/km)']

# SUV-SMALL ã®å¹³å‡ CO2 æ’å‡ºé‡ã‚’è¨ˆç®—
suv_small_co2_average = round(df_cars[cars_classes == 'SUV - SMALL']['CO2 Emissions(g/km)'].mean(), 2)
print(f"SUV-SMALL ã®å¹³å‡ CO2 æ’å‡ºé‡:{suv_small_co2_average}g/km")

#MID-SIZE ã®å¹³å‡ CO2 æ’å‡ºé‡ã‚’è¨ˆç®—
mid_size_co2_average = round(df_cars[cars_classes == 'MID-SIZE']['CO2 Emissions(g/km)'].mean(), 2)
print(f"MID-SIZE  ã®å¹³å‡ CO2 æ’å‡ºé‡:{mid_size_co2_average}g/km")
```

MID-SIZE ã®å¹³å‡ CO2 æ’å‡ºé‡ã®æ–¹ãŒSUV-SMALL ã‚ˆã‚Šã€æ’å‡ºé‡ãŒå°‘ãªã„äº‹ãŒã‚ã‹ã‚Šã¾ã™ã€‚ãƒ‡ãƒ¼ã‚¿ã‚’åˆ†æçµæœã€å°å‹SUVã®æ–¹ãŒä¸€èˆ¬ä¸­å‹è»Šã‚ˆã‚Šã‚‚1kmã‚ãŸã‚Šã®æ’å‡ºé‡ãŒå¤šããªã‚Šã¾ã—ãŸã€‚

5. ã™ã¹ã¦ã®è»Šä¸¡ã®å¹³å‡CO2æ’å‡ºé‡ã¯ã©ã‚Œãã‚‰ã„ã§ã™ã‹? ã‚¨ãƒ³ã‚¸ãƒ³ã‚µã‚¤ã‚ºãŒ2.0ãƒªãƒƒã‚¿ãƒ¼ä»¥ä¸‹ã®è»Šä¸¡ã®å¹³å‡CO2æ’å‡ºé‡ã‚‚æ±‚ã‚ã¦ä¸‹ã•ã„ã€‚

```python
#ã™ã¹ã¦ã®è»Šä¸¡ã®å¹³å‡CO2æ’å‡ºé‡ã€‚
average_co2_emission = df_cars['CO2 Emissions(g/km)'].mean()
print (f"å…¨è»Šä¸¡ã®å¹³å‡CO2æ’å‡ºé‡: {round(average_co2_emission, 2)}g/km")

#ã‚¨ãƒ³ã‚¸ãƒ³ã‚µã‚¤ã‚ºãŒ2.0ãƒªãƒƒã‚¿ãƒ¼ä»¥ä¸‹ã®è»Šä¸¡ã®å¹³å‡ CO2 æ’å‡ºé‡ã€‚
average_co2_emission_for_small_cars = df_cars[cars_engine_sizes < 2]['CO2 Emissions(g/km)'].mean()
print (f"ã‚¨ãƒ³ã‚¸ãƒ³ã‚µã‚¤ã‚ºãŒ2.0ãƒªãƒƒã‚¿ãƒ¼ä»¥ä¸‹ã®è»Šä¸¡ã®å¹³å‡CO2æ’å‡ºé‡: {round(average_co2_emission_for_small_cars,2)}g/km")
```

6. æœ€å¾Œã«æ—¥æœ¬è»Šï¼ˆãƒˆãƒ¨ã‚¿ï¼‰ã€ã‚¢ãƒ¡ãƒªã‚«è»Šï¼ˆãƒ•ã‚©ãƒ¼ãƒ‰ï¼‰ã€ã©ã¡ã‚‰ã®æ–¹ãŒã€æ’å‡ºã‚¬ã‚¹ãŒå°‘ãªã„ã®ã‹ã‚’èª¿ã¹ã¦ã¿ãŸå¾Œã€ã©ã®è»Šç¨®ãŒæœ€ã‚‚ç’°å¢ƒã«æ‚ªã„ã‹ã‚‚èª¿ã¹ã¦ã¿ã¾ã™ã€‚2ç¤¾ã®ç·åˆçš„ã«æ¯”è¼ƒã‚’ã—ã€ã©ã‚Œãã‚‰ã„ç’°å¢ƒã«å„ªã—ã„è»Šã€ã‚ã‚‹ã„ã¯ç’°å¢ƒã«æ‚ªã„è»Šã‚’ä½œã‚‹ã®ã‹ã‚’æ¯”ã¹ã¾ã™ã€‚

```python
# ãƒˆãƒ¨ã‚¿è»Šã®ã¿ã‚’å«ã‚€æ–°ã—ã„ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã‚’ä½œæˆ
toyota_cars = df_cars[df_cars['Make'] == 'TOYOTA']

# CO2æ’å‡ºé‡ã‚’å…ƒã«æœ€ã‚‚ç’°å¢ƒã«æ‚ªã„ãƒˆãƒ¨ã‚¿è»Šã‚’è¦‹ã¤ã‘ã‚‹
worst_toyota_car = toyota_cars.loc[toyota_cars['CO2 Emissions(g/km)'].idxmax()]
print(f"æœ€ã‚‚ç’°å¢ƒã«æ‚ªã„ãƒˆãƒ¨ã‚¿è»Šã¯: {worst_toyota_car['Model']},CO2æ’å‡ºé‡ã¯{worst_toyota_car['CO2 Emissions(g/km)']}(g/km)ã§ã™ã€‚")

# ãƒ•ã‚©ãƒ¼ãƒ‰è»Šã®ã¿ã‚’å«ã‚€æ–°ã—ã„ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã‚’ä½œæˆ
ford_cars = df_cars[df_cars['Make'] == 'FORD']

# CO2æ’å‡ºé‡ã‚’å…ƒã«æœ€ã‚‚ç’°å¢ƒã«æ‚ªã„ãƒ•ã‚©ãƒ¼ãƒ‰è»Šã‚’è¦‹ã¤ã‘ã‚‹
worst_ford_car = ford_cars.loc[ford_cars['CO2 Emissions(g/km)'].idxmax()]
print(f"æœ€ã‚‚ç’°å¢ƒã«æ‚ªã„ãƒ•ã‚©ãƒ¼ãƒ‰è»Šã¯: {worst_ford_car['Model']},CO2æ’å‡ºé‡ã¯{worst_ford_car['CO2 Emissions(g/km)']}(g/km)ã§ã™ã€‚")

# ãƒˆãƒ¨ã‚¿è»Šã¨ãƒ•ã‚©ãƒ¼ãƒ‰è»Šã®ã‚¨ãƒ³ã‚¸ãƒ³ã‚µã‚¤ã‚ºã¨CO2æ’å‡ºé‡ã®é–¢ä¿‚ã‚’ãƒ—ãƒ­ãƒƒãƒˆ
plt.scatter(toyota_cars['Engine Size(L)'], toyota_cars['CO2 Emissions(g/km)'], label='Toyota')
plt.scatter(ford_cars['Engine Size(L)'], ford_cars['CO2 Emissions(g/km)'], label='Ford')
plt.xlabel("Engine Size (L)")
plt.ylabel("CO2 Emissions (g/km)")
plt.legend()
plt.show()
```

```python
%%sql
SELECT 
    make,
    model,
    engine_size,
    cylinders,
    fuel_type,
    min(co2_emissions)
FROM
    vehicles
WHERE 
    engine_size < 2.0
GROUP BY 
    make
ORDER BY 
    MAKE ASC
LIMIT 30
```
