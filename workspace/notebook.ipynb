## üìñ Background
You volunteer for a public policy advocacy organization in Canada, and your colleague asked you to help her draft recommendations for guidelines on CO2 emissions rules.

## üíæ The data I

### You have access to seven years of CO2 emissions data for Canadian vehicles ([source](https://open.canada.ca/data/en/dataset/98f1a129-f628-4ce4-b24d-6f16bf24dd64#wb-auto-6)):

The data comes from the Government of Canada's open data [website](https://open.canada.ca/en).

1. What is the most expensive item your company sells? The least expensive?

```python
SELECT 
	product_name, list_price
FROM 
	products
WHERE 
	list_price = (SELECT MAX(list_price) FROM products)
UNION ALL 
SELECT 
	product_name, list_price
FROM 
	products
WHERE 
	list_price = (SELECT MIN(list_price) FROM products);
```

By answering first question, we can tell that Trek Domane SLR 9 Disc made by 2018 is the most expensive bike. This bike costs more than 11,000 dollars. the least expensive bike is Strider Classic 12 Balance Bike which made by the same year. It costs less than 100 dollars. 

2. How many different products of each category does your company sell?

```python
SELECT 
	c.category_id, category_name, sum(product_id) AS sold_per_category
FROM 
	products AS p
LEFT JOIN 
	categories AS c
ON p.category_id = c.category_id
GROUP BY 
	c.category_id, category_name
ORDER BY 
	sold_per_category DESC;
```

By answering this question, we can tell that cruisers sells bikes the most. However, runner up, children bicycles category was very close. Cyclocross bicycles and electric Bikes categories weren't sold very good. Were their categories' prices too high? Let's find out! 

## Extra Question! 
- (Do the bottom categories from the previous question charge bikes too much? what is their right price?)

```python
SELECT 
	category_name, sum(product_id) AS sold_per_category, round(avg(list_price),2) as average_category_price
FROM 
	products AS p
LEFT JOIN 
	categories AS c
ON p.category_id = c.category_id
GROUP BY 
	c.category_id, category_name
ORDER BY 
	sold_per_category DESC;
```

3. What are the top three brands with the highest average list price? The top three categories?

```python
SELECT 
	b.brand_id, b.brand_name, round(avg(list_price),2) AS average_list_price
FROM 
	products AS p
LEFT JOIN 
	brands AS b
ON p.brand_id = b.brand_id
GROUP BY 
	b.brand_id, b.brand_name
ORDER BY 
	average_list_price DESC;
```

Brand Trek is the highest average list price. Heller is the second place and so on. Does average list price change depending on the year? Let's get the answer!

4. Any other insights you found during your analysis? 
- I would love to find out the most expensive bikes per year and by brand_name. 

```python
SELECT 
	brand_name, model_year, round(max(list_price),2) AS most_expensive_by_brand_and_year
FROM 
	products AS p
LEFT JOIN 
	brands AS b
ON p.brand_id = b.brand_id
GROUP BY 
	brand_name, model_year
ORDER BY 
	most_expensive_by_brand_and_year DESC;
```

It was very interesting to find out becuase Trek bikes tend to be the most expensive bikes regardless of the year.
